# Question 3

> For the trips in November 2025, how many trips had a trip_distance of less than or equal to 1 mile?

```python
df["lpep_pickup_datetime"]=pd.to_datetime(df["lpep_pickup_datetime"])
```

```python
dfs = df[df["lpep_pickup_datetime"].between("2025-11-01","2025-12-01",inclusive="neither")]
len(dfs[dfs["trip_distance"]<=1].reset_index())
# dfs
```

# Question 4

> Which was the pick up day with the longest trip distance? Only consider trips with trip_distance less than 100 miles.

```python
df["month"]=df["lpep_pickup_datetime"].dt.month
df["day"]=df["lpep_pickup_datetime"].dt.day
df["year"]=df["lpep_pickup_datetime"].dt.year
```

```python
df[df["trip_distance"]<100].groupby(["month","day"])\
    .agg(total=("trip_distance","sum"))\
    .sort_values("total",ascending=False)
```

# Question 5

> Which was the pickup zone with the largest total_amount (sum of all trips) on November 18th, 2025?

```python
import pandas as pd

pd.set_option("display.max_columns", None)
```

```python
import pandas as pd
df_zone = pd.read_csv("./dataset/taxi_zone_lookup.csv")
df_zone=df_zone.dropna()
```

```python
df_zone
```

```python
dfs = pd.merge(df,df_zone,left_on="PULocationID",right_on="LocationID")
dfs
```

```python
dfs[
    (dfs["day"] == 18) & (dfs["month"] == 11) & (dfs["year"] == 2025) 
].groupby("Zone") \
 .agg(total=("total_amount", "sum")) \
 .sort_values("total", ascending=False)
```

# Question 6.
>For the passengers picked up in the zone named "East Harlem North" in November 2025, which was the drop off zone that had the largest tip?

```python
dfs = pd.merge(df,df_zone,left_on="DOLocationID",right_on="LocationID")
dfs
```

```python
df_name_zone = pd.merge(dfs,df_zone,left_on="PULocationID",right_on="LocationID")
```

```python
bf = df_name_zone.rename(columns={"Zone_x":"zone_drop","Zone_y":"zone_pickup"})
```

```python
bf.groupby("zone_drop").agg(total=("tip_amount","sum")).sort_values("total",ascending=False)
```
